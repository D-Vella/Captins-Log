```python
import sqlite3
from pathlib import Path
import os
```

```python
# Use Path to define the directory structure dynamically
DB_DIR_NAME = "../db"  # Go up from cli/ to project root, then into db/
DB_FILE_NAME = "application.db"

# Construct the full path for the database file
db_path = Path(DB_DIR_NAME) / DB_FILE_NAME

print(f"Target Directory: {db_path.parent}")
print(f"Target Database File: {db_path}")
```

```python
try:
    # os.makedirs creates the directory AND all necessary parent directories.
    # exist_ok=True prevents the script from crashing if the directory already exists.
    os.makedirs(db_path.parent, exist_ok=True)
    print(f"\n✅ Directory structure ensured at: {db_path.parent}")

except OSError as e:
    print(f"\n❌ Error creating directory {db_path.parent}: {e}")
    # If directory creation fails, we cannot proceed.
    exit()
```

```python
conn = sqlite3.connect(db_path)

# The cursor object is what we use to run commands
cursor = conn.cursor()

print("✅ 1. Connection established successfully!")
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS log_entry (
    id          INTEGER PRIMARY KEY,
    entry_date  DATE NOT NULL UNIQUE,
    created_at  DATETIME NOT NULL,
    updated_at  DATETIME NOT NULL
);
"""
cursor.execute(create_table_query)
print("✅ Table 'log_entry' created successfully.")
```

```python
log_seqment_query = """
CREATE TABLE IF NOT EXISTS log_segment (
    id              INTEGER PRIMARY KEY,
    log_entry_id    INTEGER NOT NULL REFERENCES log_entry(id),
    audio_filename  TEXT NOT NULL,
    duration_secs   REAL,
    raw_transcript  TEXT,
    created_at      DATETIME NOT NULL
);
"""

cursor.execute(log_seqment_query)
print("✅ Table 'log_segment' created successfully.")
```

```python
log_enrichment_query = """
CREATE TABLE IF NOT EXISTS log_enrichment (
    id              INTEGER PRIMARY KEY,
    log_entry_id    INTEGER NOT NULL REFERENCES log_entry(id),
    formatted_md    TEXT,         -- cleaned & formatted markdown
    followup_qs     TEXT,         -- JSON array of follow-up questions
    weekly_summary  TEXT,         -- populated on weekly review
    generated_at    DATETIME NOT NULL
);
"""

cursor.execute(log_enrichment_query)
print("✅ Table 'log_enrichment' created successfully.")
```

```python
conn.commit()
```
